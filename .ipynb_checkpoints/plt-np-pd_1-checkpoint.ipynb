```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid = pd.read_csv('D:\\vishal data\\csv file\\covid.csv')
```

```python
covid
```

```python
# scatter
plt.subplot(1,2,1)
plt.scatter(covid['No'], covid['active'], c = 'g', s = 30)
plt.scatter(covid['No'], covid['deaths'], c = 'r', s = 15)

plt.title('Scatter', size = 30)
plt.xlabel('state_no', size = 15, c = 'm')
plt.ylabel('active, \n deaths', size = 15, c = 'r')
plt.grid()

#pie
plt.subplot(1,2,2)
plt.pie(covid['active'], labels = covid['No'])
plt.legend(title = 'state')
plt.title('Pie_Charts', size = 30)

plt.show()
```

```python
# scatter
plt.subplot(2,2,1)
plt.scatter(covid['No'], covid['active'], c = 'g', s = 30)
plt.scatter(covid['No'], covid['deaths'], c = 'r', s = 15)

plt.grid()

plt.title('Scatter', size = 30)
plt.xlabel('state_no', size = 15, c = 'm')
plt.ylabel('active, \n deaths', size = 15, c = 'r')

#pie
plt.subplot(2,2,2)
plt.pie(covid['active'], labels = covid['No'])
plt.legend(title = 'state')
plt.title('Pie_Charts', size = 30)

#Bar
plt.subplot(2,2,3)
plt.bar(covid['No'], covid['active'], width = 0.2)
plt.xlabel('state_no', size = 15, c = 'm')
plt.ylabel('active, \n deaths', size = 15, c = 'r')

plt.show()
```

```python
option = 'option \n1 = active \n2 = deaths \n3 = dose1 \n4 = dose2 \n5 = dose3'
print(option)
check = int(input('enter your option'))

if check == 1:
    plt.scatter(covid['No'], covid['active'], c = 'g', s = 30)
    plt.title('Scatter', size = 30)
    plt.xlabel('state_no', size = 15, c = 'm')
    plt.ylabel('active', size = 15, c = 'r')
    plt.grid()

elif check == 2:
    plt.scatter(covid['No'], covid['deaths'], c = 'r', s = 15)
    plt.title('Scatter', size = 30)
    plt.xlabel('state_no', size = 15, c = 'm')
    plt.ylabel('deaths', size = 15, c = 'r')
    plt.grid()

elif check == 3:
    plt.scatter(covid['No'], covid['dose1'], c = 'r', s = 15)
    plt.title('Scatter', size = 30)
    plt.xlabel('state_no', size = 15, c = 'm')
    plt.ylabel('dose1', size = 15, c = 'r')
    plt.grid()

elif check == 4:
    plt.scatter(covid['No'], covid['dose2'], c = 'r', s = 15)
    plt.title('Scatter', size = 30)
    plt.xlabel('state_no', size = 15, c = 'm')
    plt.ylabel('dose2', size = 15, c = 'r')
    plt.grid()

elif check == 5:
    plt.scatter(covid['No'], covid['dose3'], c = 'r', s = 15)
    plt.title('Scatter', size = 30)
    plt.xlabel('state_no', size = 15, c = 'm')
    plt.ylabel('dose3', size = 15, c = 'r')
    plt.grid()


plt.show()
```

```python
Option = '1 = scatter \n2 = pie \n3 = bar'
print(Option)

user = int(input('enter your option graph'))

if user == 1:
    plt.scatter(covid['No'], covid['active'], c = 'g', s = 30)
    plt.scatter(covid['No'], covid['deaths'], c = 'r', s = 15)
    plt.title('Scatter', size = 30)
    plt.xlabel('state_no', size = 15, c = 'm')
    plt.ylabel('active, \n deaths', size = 15, c = 'r')
    plt.grid(color = 'c', ls = ':', linewidth = 1.5)
    
elif user == 2:
    plt.pie(covid['active'], labels = covid['No'])
    plt.legend(title = 'state')
    plt.title('Pie_Charts', size = 30)

elif user == 3:
    plt.bar(covid['No'], covid['active'], width = 0.5)
    plt.title('bar', size = 30)
    plt.xlabel('state_no', size = 15, c = 'm')
    plt.ylabel('active, \n deaths', size = 15, c = 'r')

plt.show()
```
