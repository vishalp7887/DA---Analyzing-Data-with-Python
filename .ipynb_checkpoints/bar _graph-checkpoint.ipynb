```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [[12,34,13,25,24],
       [28,13,35,23,26],
       [15,27,29,26,24]]
x = np.arange(5)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x + 0.00, data[0], color = 'r', width = 0.25)
ax.bar(x + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(x + 0.50, data[2], color = 'b', width = 0.25)
```
