```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr1 = np.array([1,2,3,4,5,6,7,8])
arr2 = np.array([50,30,60,40,20,40,20,30])
```

```python
plt.subplot(2,3,1)
plt.plot(arr1, arr2, marker = 'o', c = 'b', ms = 10)

plt.subplot(2,3,2)
plt.bar(arr1, arr2, color = 'k', width = 0.2)

plt.subplot(2,3,3)
plt.plot(arr1, arr2, marker = 'o', c = 'r', ms = 10)

plt.subplot(2,3,4)
plt.bar(arr1, arr2, color = 'm', width = 0.4)

plt.subplot(2,3,5)
plt.plot(arr1, arr2, marker = '*', c = 'g', ms = 10)

plt.subplot(2,3,6)
plt.bar(arr1, arr2, color = 'y', width = 0.6)

plt.show()
```
