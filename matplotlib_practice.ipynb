#### Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([5, 1, 3, 7, 6])
```

```python
# Draw a line in a diagram from position
plt.plot(arr1, arr2)
plt.show()
```

```python
# mark each point with a circle.
plt.plot(arr1, arr2, marker = 'o')
plt.show()
```

```python
# mark each point with a star.
plt.plot(arr1, arr2, marker = '*')
plt.show()
```

```python
# mark each point with a circle.
# marker|line|color
plt.plot(arr1, arr2, 'o:r')
plt.show()
```

```python
# set the size of the marker.
plt.plot(arr1, arr2, marker = 'o', ms = 15)
plt.show()
```

```python
# set the EDGE color.
plt.plot(arr1, arr2, marker = 'o', ms = 15, mec = 'r')
plt.show()
```

```python
# set the FACE color.
plt.plot(arr1, arr2, marker = 'o', ms = 15, mfc = 'r')
plt.show()
```

```python
# set the color of both the edge and the face color.
plt.plot(arr1, arr2, marker = 'o', ms = 15, mec = 'r', mfc = 'r')
plt.show()
```

```python
# use a dotted line.
plt.plot(arr1, arr2, linestyle = 'dotted')
plt.show()
```

```python
# set the line color.
plt.plot(arr1, arr2, c = 'r')
plt.show()
```

```python
# Draw two lines by specifying a plt.plot() function for each line:
plt.plot(arr1)
plt.plot(arr2)
plt.show()
```

```python
# Add labels to the x- and y-axis:
plt.plot(arr1, arr2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# Add a plot title and labels for the x- and y-axis:
plt.plot(arr1, arr2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('graph')
plt.show()
```

```python
#Set font properties for the title and labels:
plt.plot(arr1, arr2)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.xlabel('x-axis', fontdict = font1)
plt.ylabel('y-axis', fontdict = font1)
plt.title('graph', fontdict = font2)
plt.show()
```

```python
# Position the title to the left:
plt.plot(arr1, arr2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('graph', loc = 'left')
plt.show()
```

```python
# add grid lines to the plot:
plt.plot(arr1, arr2)
plt.grid()
plt.show()
```

```python
# Draw 2 plots:
x1 = np.array([1,6,3,2])
y1 = np.array([5,6,8,2])

plt.subplot(1, 2, 1)
plt.plot(x1, y1)

x2 = np.array([6,4,1,7])
y2 = np.array([8,1,7,3])

plt.subplot(1,2,2)
plt.plot(x2,y2)

plt.show()
```

```python
# 2 plots, with titles:
x1 = np.array([1,6,3,2])
y1 = np.array([5,6,8,2])

plt.subplot(1, 2, 1)
plt.plot(x1, y1)
plt.title('first')

x2 = np.array([6,4,1,7])
y2 = np.array([8,1,7,3])

plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.title('second')

plt.show()
```

```python
# Add a title for the enre figure:
x1 = np.array([1,6,3,2])
y1 = np.array([5,6,8,2])

plt.subplot(1, 2, 1)
plt.plot(x1, y1)
plt.title('first')

x2 = np.array([6,4,1,7])
y2 = np.array([8,1,7,3])

plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.title('second')

plt.suptitle('graph')
plt.show()
```

```python
# a simple scatter plot:
plt.scatter(arr1, arr2)
plt.show()
```

```python
# Draw two plots on the same figure:
plt.scatter(arr1, arr2)
plt.scatter(x1, y1)
plt.show()
```

```python
# creating bar graph:
b1 = np.array(['a', 'b', 'c', 'd'])
b2 = np.array([5,8,3,5])

plt.bar(b1, b2)
plt.show()
```

```python
# horizontal bar graph:
b1 = np.array(['a', 'b', 'c', 'd'])
b2 = np.array([5,8,3,5])

plt.barh(b1, b2)
plt.show()
```

```python
# a simple histogram 
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

```python
# a simple pie chart
p = np.array([5,8,3,9,2,4])
plt.pie(p)
plt.show()
```

```python
# pie chart labels:
p = np.array([5,8,3,9,2,4])
l = ['a', 'b', 'c', 'd', 'e', 'f']
plt.pie(p, labels = l)
plt.show()
```

```python
# Start the first wedge at 90 degrees:
p = np.array([5,8,3,9,2,4])
l = ['a', 'b', 'c', 'd', 'e', 'f']
plt.pie(p, labels = l, startangle = 90)
plt.show()
```
