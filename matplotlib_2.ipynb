```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = [1,3,4,5]
b = [3,6,7,3]

f1 = {'family' : 'serif', 'color' : 'blue', 'size' : 35}
f2 = {'family' : 'serif', 'color' : 'red', 'size' : 20}

plt.title('graph', fontdict = f1)
plt.xlabel('x_axis', fontdict = f2)
plt.ylabel('y_axis', fontdict = f2)

plt.plot(a, b, 'o')
plt.show()
```

```python
covid = pd.read_csv('D:\\vishal data\\csv file\\covid.csv')
```

```python
covid1 = plt.plot(covid['No'], covid['active'],'o',  covid['deaths'], 'o', ms = 5)

plt.title('graph', size = 30)
plt.xlabel('state_no', size = 20)
plt.ylabel('active', size = 20)

plt.show()
```

```python
covid2 = plt.plot(covid['No'], covid['active'])
plt.title('graph', loc = 'left')
plt.xlabel('x-axis', loc = 'right')
plt.ylabel('y-axis')

plt.grid()

plt.show()
```

```python
#subplot

plt.subplot(2,2,1)
plt.plot(covid['No'], covid['deaths'])
plt.title('deaths')

plt.subplot(2,2,2)
plt.plot(covid['No'], covid['active'])
plt.title('active')

plt.subplot(2,2,3)
plt.plot(covid['No'], covid['dose1'])
plt.title('dose1')

plt.subplot(2,2,4)
plt.plot(covid['No'], covid['dose2'])
plt.title('dose2')

plt.show()
```

```python
plt.scatter(covid['No'], covid['active'], c = 'r')
plt.scatter(covid['No'], covid['deaths'], c = 'b')

plt.show()
```
